```python
import keras

from keras.preprocessing import image
from keras.applications.inception_v3 import preprocess_input,decode_predictions

import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import time

model = keras.applications.inception_v3.InceptionV3(include_top=True, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)
```

```python
model.summary()
model.get_config()
```

```python
img_path = '../dataset/animal/cat.jpg'
img = image.load_img(img_path, target_size=(299, 299))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

begin = time.clock()
preds = model.predict(x)
print('Predicted:', decode_predictions(preds))
print ('Time elapsed: %.0f' % (time.clock() - begin))
```
